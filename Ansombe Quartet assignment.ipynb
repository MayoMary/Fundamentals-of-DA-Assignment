## Importing the required libraries via the Pandas package which contains fuunctions to assist in analysing data. 
Pandas is intended for "real world data analysis in Python." It is an open source data analysis/manipulation tool
available in all languages. 
Command\functions start as "pd." and it reads the data file.

```python
import pandas as pd
```

```python
# The 4 sources of data (quartet) representing the 4 days which need to be 'read' by the Pandas package, hence.".pd". 
# They are then labeled data1 to data4.

data1 = pd.read_csv("day1.csv")
data2 = pd.read_csv("day2.csv")
data3 = pd.read_csv("day3.csv")
data4 = pd.read_csv("day4.csv")

print("This is first dataset")
print(data1)
print("This is second dataset")
print(data2)
print("This is third dataset")
print(data3)
print("This is fourth dataset")
print(data4)
```

```python
# Use of "space and \n" starts a new row after each calculation to make reading the results much easier.
# The "axis=0" in the command is to show whether the data is to be analysed columnwise or rowwise. 
# For this dataset it is row wise, therefore "axis=0". Columnwise would be axis=1.
# Reference : https://pandas.pydata.org/pandas-docs/stable/index.html

print("Mean \n",data1.mean(axis=0))
print("Min \n",data1.min(axis=0))
print("Max \n",data1.max(axis=0))
print("Median \n",data1.median(axis=0))
print("Std Dev. \n",data1.std(axis=0))
```

```python
print("Mean \n",data2.mean(axis=0))
print("Min \n",data2.min(axis=0))
print("Max \n",data2.max(axis=0))
print("Median \n",data2.median(axis=0))
print("Std Dev. \n",data2.std(axis=0))
```

```python
print("Mean \n",data3.mean(axis=0))
print("Min \n",data3.min(axis=0))
print("Max \n",data3.max(axis=0))
print("Median \n",data3.median(axis=0))
print("Std Dev. \n",data3.std(axis=0))
```

```python
print("Mean \n",data4.mean(axis=0))
print("Min \n",data4.min(axis=0))
print("Max \n",data4.max(axis=0))
print("Median \n",data4.median(axis=0))
print("Std Dev. \n",data4.std(axis=0))
```

## Data analysis
Having analysed the data, it is however visually hard to compare the results on screen when there are four data sets listed. 
After searching for a more compact analysis Panda function - ".describe()" allows the dataset to be summarised concisely and contains the min, max, mean, standard deviation and other analytical results of the data
Reference : https://pandas.pydata.org/pandas-docs/stable/index.html


```python
summary1=data1.describe()
summary2=data2.describe()
summary3=data3.describe()
summary4=data4.describe()

print(summary1,"\n",summary2,"\n", summary3,"\n", summary4)

# Use of the "List" function allows the data to be shown with the analysis of all 4 datasets in a list format shown one after the other.

list_of_summaries = [summary1,summary2,summary3,summary4]

# Use of "concat" function to display all the 4 datasets together
result=pd.concat(list_of_summaries)
```

```python
## The results
result
```

## The result of the analysis:
That the statistical values for the mean, minimum and maximum are the same for most of the 'x' datasets, for example the max is 14 for x1, x2 and x3. Mean of x is 9 and mean of y is 7.50 for all 4 datasets.
